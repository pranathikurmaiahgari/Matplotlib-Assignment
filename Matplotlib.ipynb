Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library for creating static, interactive, and animated visualizations in a wide range of formats and styles. It provides a wide variety of plotting capabilities, making it a valuable tool for data visualization and exploration. Matplotlib is used extensively in data analysis, scientific research, engineering, and various other fields where visualizing data is important.

The library allows you to create a variety of plots, charts, and graphs to effectively convey information. Some of the key features of Matplotlib include support for various plot types, customization of plot appearance, and the ability to create complex visualizations.

Here are five common plots that can be created using the pyplot module of Matplotlib:

Line Plot: A line plot (also known as a line chart) is used to display data points connected by straight lines. It's commonly used to show trends over time or to visualize the relationship between two variables.

Bar Plot: A bar plot (or bar chart) is used to represent categorical data with rectangular bars. Each bar represents a category, and the length of the bar corresponds to the quantity or value associated with that category.

Histogram: A histogram is used to visualize the distribution of a continuous dataset. It represents data in intervals or bins and displays the frequency or count of data points falling within each bin.

Scatter Plot: A scatter plot is used to display individual data points as markers on a two-dimensional plane. It's often used to show the relationship between two continuous variables.

Pie Chart: A pie chart is a circular graph divided into sectors, each representing a proportion of the whole. It's commonly used to show how a single data point or category contributes to the whole dataset.




Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of data visualization that displays individual data points as markers on a two-dimensional plane. Each data point is represented by a marker, such as a point or a small circle, and the position of the marker on the plane is determined by the values of two variables (typically the x-axis and y-axis). Scatter plots are particularly useful for visualizing the relationship between two continuous variables and identifying patterns, trends, clusters, or outliers in the data.

Scatter plots are widely used in various fields, including statistics, data analysis, and scientific research. They can help answer questions such as:

Are there any patterns or trends between two variables?
Is there a correlation between the two variables? If so, is it positive or negative?
Are there any outliers that deviate significantly from the general pattern?
In a scatter plot:

The horizontal axis represents one variable (often referred to as the independent variable or x-variable).
The vertical axis represents the other variable (often referred to as the dependent variable or y-variable).
Each data point is represented by a marker at the intersection of the corresponding x and y values.



```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
```

The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to arrange multiple plots in a grid-like layout, making it convenient to compare different datasets or visualizations side by side.

```python
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(3,3,1)
plt.plot(y1)
plt.title("Line 1")
plt.subplot(3,3,2)
plt.plot(y2)
plt.title("Line 2")
plt.subplot(3,3,3)
plt.plot(y3)
plt.title("Line 3")
plt.subplot(3,3,4)
plt.plot(y4)
plt.title("Line 4")
```

A bar plot, also known as a bar chart, is a graphical representation of categorical data using rectangular bars. Each bar represents a category, and the length of the bar corresponds to the value or quantity associated with that category. Bar plots are commonly used to compare different categories or to display data across different groups or time periods.

Bar plots are particularly useful for:

Comparing values between different categories.
Visualizing distributions or frequencies of categorical data.
Showing trends or patterns within a dataset.
You can create both vertical and horizontal bar plots, depending on the orientation of the bars.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It summarizes the minimum, first quartile (25th percentile), median (second quartile or 50th percentile), third quartile (75th percentile), and maximum values of a dataset. The box plot provides insights into the central tendency, spread, and potential outliers of the data distribution.

Box plots are useful for:

Identifying the central tendency of the data (median).
Understanding the spread or variability of the data (interquartile range, IQR).
Detecting potential outliers in the data.
Comparing the distributions of different datasets.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data=[box1,box2]
plt.boxplot(data)
```
